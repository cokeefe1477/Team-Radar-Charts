```python
def TeamRadarCharts(filename, teamname, fileexport, FD_feats, W_feats, AM_feats, MF_feats, OB_feats, CB_feats, GK_feats):
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    from soccerplots import radar_chart
    
    team_data = pd.read_csv(filename)
    
    forward_group_lst = ["ST", "CF", "RF", "LF", "RS", "LS"]
    winger_group_lst = ["RW", "LW", "RM", "LM"] 
    attmid_group_lst = ["CAM", "RAM", "LAM"]
    midfield_group_lst = ["CM", "CDM", "RCM", "LCM", "RDM", "LDM"]
    outback_group_lst = ["LB", "RB", "LWB", "RWB"]
    cenback_group_lst = ["CB", "LCB", "RCB"]
    
    pos_group_lst = []
    for index, row in team_data.iterrows():
        if row["Pos"] in forward_group_lst:
            pos_group_lst.append("forward")
        elif row["Pos"] in winger_group_lst:
            pos_group_lst.append("winger")
        elif row["Pos"] in attmid_group_lst:
            pos_group_lst.append("attmid")
        elif row["Pos"] in midfield_group_lst:
            pos_group_lst.append("midfield")
        elif row["Pos"] in outback_group_lst:
            pos_group_lst.append("outback")
        elif row["Pos"] in cenback_group_lst:
            pos_group_lst.append("cenback")
        elif row["Pos"] == "GK":
            pos_group_lst.append("goalkeeper")
    team_data["Pos_Group"] = pos_group_lst
    
    team_dict = team_data.to_dict(orient = "list")
    FD_dict = {}
    W_dict = {}
    AM_dict = {}
    MF_dict = {}
    OB_dict = {}
    CB_dict = {}
    GK_dict = {}
    
    
    for i in range(len(team_dict["Player"])):
        values_lst = []
        if team_data["Pos_Group"][i] == "forward":
            for j in range(len(FD_feats)):
                values_lst.append(team_data[FD_feats[j]][i])
                FD_dict[team_data["Player"][i]] = values_lst
        if team_data["Pos_Group"][i] == "winger":
            for j in range(len(W_feats)):
                values_lst.append(team_data[W_feats[j]][i])
                W_dict[team_data["Player"][i]] = values_lst
        if team_data["Pos_Group"][i] == "attmid":
            for j in range(len(AM_feats)):
                values_lst.append(team_data[AM_feats[j]][i])
                AM_dict[team_data["Player"][i]] = values_lst
        if team_data["Pos_Group"][i] == "midfield":
            for j in range(len(MF_feats)):
                values_lst.append(team_data[MF_feats[j]][i])
                MF_dict[team_data["Player"][i]] = values_lst
        if team_data["Pos_Group"][i] == "outback":
            for j in range(len(OB_feats)):
                values_lst.append(team_data[OB_feats[j]][i])
                OB_dict[team_data["Player"][i]] = values_lst
        if team_data["Pos_Group"][i] == "cenback":
            for j in range(len(CB_feats)):
                values_lst.append(team_data[CB_feats[j]][i])
                CB_dict[team_data["Player"][i]] = values_lst
        if team_data["Pos_Group"][i] == "goalkeeper":
            for j in range(len(GK_feats)):
                values_lst.append(team_data[GK_feats[j]][i])
                GK_dict[team_data["Player"][i]] = values_lst
    
    for key in FD_dict:
        FD_feat = FD_dict[key]
        for i in range(len(FD_feat)):
            if np.isnan(FD_feat[i]):
                FD_feat[i] = np.nan_to_num(FD_feat[i])           
    for key in W_dict:
        W_feat = W_dict[key]
        for i in range(len(W_feat)):
            if np. isnan(W_feat[i]):
                W_feat[i] = np.nan_to_num(W_feat[i])     
    for key in AM_dict:
        AM_feat = AM_dict[key]
        for i in range(len(AM_feat)):
            if np. isnan(AM_feat[i]):
                AM_feat[i] = np.nan_to_num(AM_feat[i])     
    for key in MF_dict:
        MF_feat = MF_dict[key]
        for i in range(len(MF_feat)):
            if np. isnan(MF_feat[i]):
                 MF_feat[i] = np.nan_to_num(MF_feat[i])     
    for key in OB_dict:
        OB_feat = OB_dict[key]
        for i in range(len(OB_feat)):
            if np. isnan(OB_feat[i]):
                OB_feat[i] = np.nan_to_num(OB_feat[i])       
    for key in CB_dict:
        CB_feat = CB_dict[key]
        for i in range(len(CB_feat)):
            if np. isnan(CB_feat[i]):
                CB_feat[i] = np.nan_to_num(CB_feat[i])       
    for key in GK_dict:
        GK_feat = GK_dict[key]
        for i in range(len(GK_feat)):
            if np. isnan(GK_feat[i]):
                GK_feat[i] = np.nan_to_num(GK_feat[i])
    
    FD_ranges = [[] for i in range(9)]
    W_ranges = [[] for i in range(9)]
    AM_ranges = [[] for i in range(9)]
    MF_ranges = [[] for i in range(11)]
    OB_ranges = [[] for i in range(9)]
    CB_ranges = [[] for i in range(9)]
    GK_ranges = [[] for i in range(4)]

    for key in FD_dict:
        FD_feats1 = FD_dict[key]
        for i in range(len(FD_feats1)):
            FD_ranges[i].append(FD_feats1[i])        
    for i in range(len(FD_ranges)):
        minimum = (0.95 * min(FD_ranges[i]))
        maximum = (1.05*max(FD_ranges[i]))
        range_values = (minimum, maximum)
        FD_ranges[i] = range_values 
    for key in W_dict:
        W_feats1 = W_dict[key]
        for i in range(len(W_feats1)):
            W_ranges[i].append(W_feats1[i])         
    for i in range(len(W_ranges)):
        minimum = (0.95 * min(W_ranges[i]))
        maximum = (1.05*max(W_ranges[i]))
        range_values = (minimum, maximum)
        W_ranges[i] = range_values    
    for key in AM_dict:
        AM_feats1 = AM_dict[key]
        for i in range(len(AM_feats1)):
            AM_ranges[i].append(AM_feats1[i])  
    for i in range(len(AM_ranges)):
        minimum = (0.95 * min(AM_ranges[i]))
        maximum = (1.05*max(AM_ranges[i]))
        range_values = (minimum, maximum)
        AM_ranges[i] = range_values 
    for key in MF_dict:
        MF_feats1 = MF_dict[key]
        for i in range(len(MF_feats1)):
            MF_ranges[i].append(MF_feats1[i])       
    for i in range(len(MF_ranges)):
        minimum = (0.95 * min(MF_ranges[i]))
        maximum = (1.05*max(MF_ranges[i]))
        range_values = (minimum, maximum)
        MF_ranges[i] = range_values
    for key in OB_dict:
        OB_feats1 = OB_dict[key]
        for i in range(len(OB_feats1)):
            OB_ranges[i].append(OB_feats1[i])   
    for i in range(len(OB_ranges)):
        minimum = (0.95 * min(OB_ranges[i]))
        maximum = (1.05*max(OB_ranges[i]))
        range_values = (minimum, maximum)
        OB_ranges[i] = range_values  
    for key in CB_dict:
        CB_feats1 = CB_dict[key]
        for i in range(len(CB_feats1)):
            CB_ranges[i].append(CB_feats1[i])   
    for i in range(len(CB_ranges)):
        minimum = (0.95 * min(CB_ranges[i]))
        maximum = (1.05*max(CB_ranges[i]))
        range_values = (minimum, maximum)
        CB_ranges[i] = range_values 
    for key in GK_dict:
        GK_feats1 = GK_dict[key]
        for i in range(len(GK_feats1)):
            GK_ranges[i].append(GK_feats1[i])   
    for i in range(len(GK_ranges)):
        minimum = (0.95 * min(GK_ranges[i]))
        maximum = (1.05*max(GK_ranges[i]))
        range_values = (minimum, maximum)
        GK_ranges[i] = range_values 
        
    import matplotlib.backends.backend_pdf
    from matplotlib.backends.backend_pdf import PdfPages
    
    with PdfPages(fileexport) as export_pdf:
        for key in FD_dict:
            title = dict(title_name = key, subtitle_name = teamname + " Forward", title_fontsize = 18, subtitle_fontsize = 13, title_color = "#ff0000", subtitle_color = "#ff0000")
            fig, ax = radar_chart.plot_radar(ranges = FD_ranges, params = FD_features, values = FD_dict[key], radar_color = ["#ff0000", "#ffffff"], dpi = 500, title = title)
            export_pdf.savefig()
            plt.close()
        for key in W_dict:
            title = dict(title_name = key, subtitle_name = teamname + " Winger/Wide Mid", title_fontsize = 18, subtitle_fontsize = 13, title_color = "#ff0000", subtitle_color = "#ff0000")
            fig, ax = radar_chart.plot_radar(ranges = W_ranges, params = W_features, values = W_dict[key], radar_color = ["#ff0000", "#ffffff"], dpi = 500, title = title)
            export_pdf.savefig()
            plt.close()
        for key in AM_dict:
            title = dict(title_name = key, subtitle_name = teamname + " Attacking Mid", title_fontsize = 18, subtitle_fontsize = 13, title_color = "#ff0000", subtitle_color = "#ff0000")
            fig, ax = radar_chart.plot_radar(ranges = AM_ranges, params = AM_features, values = AM_dict[key], radar_color = ["#ff0000", "#ffffff"], dpi = 500, title = title)
            export_pdf.savefig()
            plt.close()
        for key in MF_dict:
            title = dict(title_name = key, subtitle_name = teamname + " Central Mid", title_fontsize = 18, subtitle_fontsize = 13, title_color = "#ff0000", subtitle_color = "#ff0000")
            fig, ax = radar_chart.plot_radar(ranges = MF_ranges, params = MF_features, values = MF_dict[key], radar_color = ["#ff0000", "#ffffff"], dpi = 500, title = title)
            export_pdf.savefig()
            plt.close()
        for key in OB_dict:
            title = dict(title_name = key, subtitle_name = teamname + " Outside Back", title_fontsize = 18, subtitle_fontsize = 13, title_color = "#ff0000", subtitle_color = "#ff0000")
            fig, ax = radar_chart.plot_radar(ranges = OB_ranges, params = OB_features, values = OB_dict[key], radar_color = ["#ff0000", "#ffffff"], dpi = 500, title = title)
            export_pdf.savefig()
            plt.close()
        for key in CB_dict:
            title = dict(title_name = key, subtitle_name = teamname + " Center Back", title_fontsize = 18, subtitle_fontsize = 13, title_color = "#ff0000", subtitle_color = "#ff0000")
            fig, ax = radar_chart.plot_radar(ranges = CB_ranges, params = CB_features, values = CB_dict[key], radar_color = ["#ff0000", "#ffffff"], dpi = 500, title = title)
            export_pdf.savefig()
            plt.close()
        for key in GK_dict:
            title = dict(title_name = key, subtitle_name = teamname + " Goalkeeper", title_fontsize = 18, subtitle_fontsize = 13, title_color = "#ff0000", subtitle_color = "#ff0000")
            fig, ax = radar_chart.plot_radar(ranges = GK_ranges, params = GK_features, values = GK_dict[key], radar_color = ["#ff0000", "#ffffff"], dpi = 500, title = title)
            export_pdf.savefig()
            plt.close()
    
    print("Done")
```
